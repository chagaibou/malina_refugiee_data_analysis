Comment le nombre de réfugiés a-t-il évolué au fil des années ?

```python
refugiee_population_over_time = refugiee_dataset.groupby(['reference_period_start_year'])['population'].sum().reset_index()
```

```python
refugiee_population_over_time
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=refugiee_population_over_time, x='reference_period_start_year', y='population',color='mediumseagreen',marker='o')
plt.xlabel("reference_period_start")
plt.ylabel("Population")
plt.title("Évolution de la population réfugiée du Mali au fil du temps")
plt.xticks(rotation=45)
plt.show()
```

Comment la répartition des réfugiés par tranche d'âge a-t-elle évolué dans le temps ?

```python
repartiton_refugies_by_age_range = refugiee_dataset.groupby(['age_range', 'reference_period_start_year'])['population'].sum().reset_index()
```

```python

pivot_age_year = repartiton_refugies_by_age_range.pivot(index='reference_period_start_year', columns='age_range', values='population')

# Tracer les courbes
plt.figure(figsize=(14,10))
pivot_age_year.plot()
plt.title("Évolution du nombre de réfugiés par tranche d'âge (par année)")
plt.xlabel("Année")
plt.ylabel("Nombre de réfugiés")
plt.legend(title="Tranche d'âge", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

Les pays figurant dans les plans d’intervention humanitaire (HRP) ont-ils plus de réfugiés ?

```python
refugiee_origin_with_hrp = refugiee_dataset[refugiee_dataset['origin_has_hrp']=='True'].groupby(['origin_location_name'])['population'].sum().sort_values(ascending=False).reset_index()
refugiee_origin_with_not_hrp = refugiee_dataset[refugiee_dataset['origin_has_hrp']=='False'].groupby(['origin_location_name'])['population'].sum().sort_values(ascending=False).reset_index()
```

```python
print(f'la population pour les pays avec de es plans d’intervention humanitaire (HRP) est de {refugiee_origin_with_hrp["population"].sum()}')
print(f'la population pour les pays sans de es plans d’intervention humanitaire (HRP) est de {refugiee_origin_with_not_hrp["population"].sum()}')
```

Les pays figurant dans les plans d’intervention humanitaire (HRP) ont-ils plus de réfugiés ?

```python
populations = {
    'Avec HRP': refugiee_origin_with_hrp["population"].sum(),
    'Sans HRP': refugiee_origin_with_not_hrp["population"].sum()
}
categories = list(populations.keys())
values = list(populations.values())

plt.figure(figsize=(8, 6))
plt.bar(categories, values, color=['skyblue', 'lightcoral'])
plt.xlabel("Pays d'origine")
plt.ylabel("Population totale")
plt.title("Comparaison de la population réfugiée : Avec et Sans HRP")
plt.show()
```

Les pays inclus dans le GHO fournissent-ils un nombre de réfugiés significatif par rapport aux autres ?

```python
refugiee_origin_with_gho= refugiee_dataset[refugiee_dataset['origin_in_gho']=='True'].groupby(['origin_location_name'])['population'].sum().sort_values(ascending=False).reset_index()
refugiee_origin_with_not_gho = refugiee_dataset[refugiee_dataset['origin_in_gho']=='False'].groupby(['origin_location_name'])['population'].sum().sort_values(ascending=False).reset_index()
```

```python
print(f'la population pour les pays inclus dans le gho (HRP) est de {refugiee_origin_with_gho["population"].sum()}')
print(f'la population pour les pays non inclus dans le gho (HRP) est de {refugiee_origin_with_not_gho["population"].sum()}')
```
